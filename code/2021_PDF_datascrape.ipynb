```python
import pandas as pd
import tabula
from tabula.io import read_pdf
```

```python
# read tables from the PDF file
pdf_path = "Final-2021-SUPHR-AR.pdf"
tables = tabula.io.read_pdf(pdf_path, pages=[14,16,25,27,37,38,39,40,41,42,43,44], multiple_tables=True)
```

```python
# display each of the dataframes
for df in tables:
    print(df.size)
    print(df)
```

```python
df = pd.concat(tables)
```

```python
df.to_csv('2021dfs.csv', index=False)
```
